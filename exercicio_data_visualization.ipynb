# Atividade 1

criar um gráfico de linhas comparando os números de imigrantes do Brasil e Argentina para o Canadá, que são os maiores países da América do Sul.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv')
df
```

```python
df.set_index('País', inplace = True)
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
df_comparacao = df_comparacao.T
df_comparacao
```

```python
plt.plot(df_comparacao['Brasil'], label = 'Brasil')
plt.plot(df_comparacao['Argentina'], label = 'Argentina')
plt.title('Comparação de imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

```python
# Ambos os países tiveram números próximos durante as décadas de 80 e 90, com Argentina sofrendo queda por volta do ano de 2005, enquanto o Brasil teve aumento no número de imigrantes a partir do mesmo ano.
```

# Atividade 2

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
dados_finais = pd.DataFrame(vendas_2022, index = lojas)
dados_finais
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas por loja')

axs[0,0].plot(dados_finais.loc['A'])
axs[0,0].set_title('Loja A')

axs[0,1].plot(dados_finais.loc['B'])
axs[0,1].set_title('Loja B')

axs[1,0].plot(dados_finais.loc['C'])
axs[1,0].set_title('Loja C')

axs[1,1].plot(dados_finais.loc['D'])
axs[1,1].set_title('Loja D')

for ax in axs.flat:
  ax.set_ylabel('Número de vendas')

for ax in axs.ravel():
  ax.set_ylim(0, 500)

plt.show()
```

# Atividade 3

Customizando os gráficos

```python
fig, axs = plt.subplots(2,2, figsize = (12,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas por loja', fontsize = 18)

axs[0,0].plot(dados_finais.loc['A'], color = 'green')
axs[0,0].set_title('Loja A')

axs[0,1].plot(dados_finais.loc['B'], color = 'red')
axs[0,1].set_title('Loja B')

axs[1,0].plot(dados_finais.loc['C'], color = 'pink')
axs[1,0].set_title('Loja C')

axs[1,1].plot(dados_finais.loc['D'], color = 'purple')
axs[1,1].set_title('Loja D')

for ax in axs.flat:
  ax.set_ylabel('Número de vendas', fontsize = 12)
  ax.set_xlabel('Meses', fontsize = 12)
  ax.xaxis.set_tick_params(labelsize = 10)
  ax.yaxis.set_tick_params(labelsize = 10)

for ax in axs.ravel():
  ax.set_ylim(0, 500)

plt.show()
```

Nesta etapa, seu desafio é criar uma figura contendo as tendências de imigração dos 4 maiores países da América latina: Brasil, Argentina, Peru e Colômbia. Através dessa criação você pode explorar diversas possibilidades e reconhecer de forma atrativa o seu processo de desenvolvimento.E não nos esqueçamos das orientações! Essa figura precisa ter uma linha para cada país, título, rótulos nos eixos, cores apropriadas, um tema da biblioteca Seaborn e legenda. Por isso, pense nas questões de acessibilidade, como tamanho das fontes e espessura das linhas. É importante escolher cores adequadas que não causem cansaço visual ou dificultem a leitura das informações. Além disso, o tamanho das fontes deve ser legível o suficiente para que as pessoas possam interpretar os dados com facilidade.

```python
import seaborn as sns
```

```python
america_sul = df.query('Região == "América do Sul"').sort_values('Total', ascending = False).head(4)
america_sul
```

```python
fig, ax = plt.subplots(figsize = (8,4))
ax = sns.barplot(data= america_sul, x = 'Total', y = america_sul.index, orient = 'h', palette = 'rocket')

ax.set_title('Países da América do Sul com maior número de imigração para o Canadá \n 1980 a 2013', fontsize = 14)
ax.set_xlabel('Número de imigrantes', fontsize = 12)
ax.set_ylabel('')

plt.show()
```

```python
# Correção com gráfico de linhas

sns.set_theme()

fig, ax = plt.subplots(figsize = (10,5))
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax = sns.lineplot(america_sul.loc['Guiana', anos], label = 'Guiana', lw = 1.5)
ax = sns.lineplot(america_sul.loc['Colômbia', anos], label = 'Guiana', lw = 1.5)
ax = sns.lineplot(america_sul.loc['Peru', anos], label = 'Guiana', lw = 1.5)
ax = sns.lineplot(america_sul.loc['Brasil', anos], label = 'Guiana', lw = 1.5)

ax.set_title('Países da América do Sul com maior número de imigração para o Canadá \n 1980 a 2013', fontsize = 14)
ax.set_xlabel('Anos', fontsize = 12)
ax.set_ylabel('Número de imigrantes', fontsize = 12)

ax.legend(title = 'Países', loc = 'upper right', bbox_to_anchor = (1.18, 1.02) )

plt.show()
```

# Atividade 5

```python
df.head()
```

```python
brasil = df.loc['Brasil', anos]
brasil.head()
```

```python
brasil_dict =  {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil.head()
```

```python
argentina = df.loc['Argentina', anos]
argentina.head()
```

```python
argentina_dict = {'ano': argentina.index.tolist(), 'imigrante': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
dados_argentina.head()
```

```python
dados_brasil['ano'] = dados_brasil['ano'].astype(int)
dados_argentina['ano'] = dados_argentina['ano'].astype(int)
```

```python
import plotly.graph_objs as go
```

```python
# Criar figura
fig = go.Figure()

# Adicionando as linhas de dados

fig.add_trace(go.Scatter(x = [dados_brasil['ano'].iloc[0]], y = [dados_brasil['imigrantes'].iloc[0]], mode = 'lines', name = 'Imigrantes Brasil', line = dict(width = 4)))
fig.add_trace(go.Scatter(x = [dados_argentina['ano'].iloc[0]], y = [dados_argentina['imigrante'].iloc[0]], mode = 'lines', name = 'Imigrantes Argentina', line = dict(width = 4)))

# Layout

fig.update_layout(
    title = dict(
        text = '<b>Imigração do Brasil e Argentina para o Canadá de 1980 a 2013</b>',
        x = 0.1,
        font = dict(size = 18)
    ),

    xaxis = dict(range = [1980, 2013], autorange = False, title = '<b>Ano</b>'),
    yaxis = dict(range = [0, 3000], autorange = False, title = '<b>Número de imigrantes</b>'),

    #botão
    updatemenus = [
        dict(
            type = 'buttons',
            showactive = False,
            buttons = [{
                'label': 'Play',
                'method': 'animate',
                'args': [None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
            }]
        )],
    width = 1000,
    height = 500
)

frames = []

for i in range(len(dados_brasil)):

  frames_data = [
      go.Scatter(x = dados_brasil['ano'].iloc[:i + 1], y = dados_brasil['imigrantes'].iloc[:i + 1]),
      go.Scatter(x = dados_argentina['ano'].iloc[:i + 1], y = dados_argentina['imigrante'].iloc[:i + 1])
      ]

  frame = go.Frame(data = frames_data)
  frames.append(frame)

fig.frames = frames

fig.show()
```
